## **The Label Encoding to "Make" and "Model" Columns**

```python
# the label encoding
from sklearn.preprocessing import LabelEncoder

lenc = LabelEncoder()

data['Make']=lenc.fit_transform(data['Make'])
data['Model']=lenc.fit_transform(data['Model'])
```

```python
data
```

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()

data[['Mileage']] = mms.fit_transform(data[['Mileage']])
```

```python
data
```

```python
x = data.drop('Price', axis=1) #dropping the target variable
y = data['Price'] # adding the target variable
```

```python
# partitioning
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=11)
```

```python
print(xtrain)
print(ytrain)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor

models = []
models.append(['XGB',XGBRegressor()])
models.append(['KNeigbors',KNeighborsRegressor()])
models.append(['DecisionTree',DecisionTreeRegressor()])
models.append(['RandomForest',RandomForestRegressor()])
models.append(['LightGBM',LGBMRegressor()])
```

```python
models
```

```python
from sklearn.metrics import r2_score
```

```python
l1 = []
for m in range(len(models)):
    l2 = []
    model = models[m][1]
    model.fit(xtrain,ytrain) # fitting the model with training set
    y_pred = model.predict(xtest) # making prediction with test set
    
    l2.append(models[m][0])
    l2.append(r2_score(ytest,y_pred))
    l1.append(l2) # listing the model rsquared score
    print(" l1 for m",m,"= ",l1)
```

```python
df = pd.DataFrame(l1,columns=['Model','r2_score']) # adding in a new dataframe
df.sort_values(by=['r2_score'],inplace=True,ascending=True) # sorting by rsquared-score
df
```
